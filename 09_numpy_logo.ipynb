# Numpy Logo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
try:
    logo = np.load('numpy-logo.npy')
    #display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.title('Numpy Logo')
    plt.imshow(logo)
    plt.axis('off')

    dark_logo = 1 - logo
    plt.subplot(122)
    plt.title('Numpy Dark Logo')
    plt.imshow(dark_logo)
    plt.axis('off')


except FileNotFoundError:
    print("File not found")
```
